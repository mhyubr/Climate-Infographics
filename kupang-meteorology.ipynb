# Data Preprocessing

**Boxplot Distribusi semua fitur**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_meteorologi, color='lightblue')
plt.title('\nBoxplot Dataset\n')
plt.xlabel('Curah Hujan')
plt.xticks(rotation=90)
plt.show()
```

```python
data_meteorologi[data_meteorologi['curah hujan'] > 2000][['Tanggal', 'curah hujan']]
```

**Terdapat 109 baris anomali pada data variabel curah hujan yaitu curah hujan 8888 sehingga sebelum di analisis agar anomali tersebut menjadi NaN dan tidak menghapus barisnya agar tidak kehilangan informasi pada data tanggal tertentu**

```python
data_meteorologi['curah hujan'].replace(8888, np.NaN, inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_meteorologi, color='lightblue')
plt.title('\nBoxplot Dataset\n')
plt.xlabel('Curah Hujan')
plt.xticks(rotation=90)
plt.show()
```

**Menganalisis penamaan kolom pada variabel temperature max dan min**

```python
pd.DataFrame(data_meteorologi['Temperature max'] >= data_meteorologi['Temperature min']).value_counts()
```

**Mengganti nama kolom Temperature max dan min yang terbalik**

```python
data_meteorologi = data_meteorologi.rename(columns={'Temperature max': 'Temperature min', 'Temperature min': 'Temperature max'})
pd.DataFrame(data_meteorologi['Temperature max'] >= data_meteorologi['Temperature min']).value_counts()
```

**Akan tetapi masih terdapat beberapa data yang bernilai NaN atau kosong pada kedua variabel tersebut, agar tetap menjaga keaslian data kita tidak perlu melakukan impute missing value**

```python
data_meteorologi[~(data_meteorologi['Temperature max'] >= data_meteorologi['Temperature min'])][['Temperature max', 'Temperature min']]
```

# Univariate Analysis

**Menganalisis Arah Angin Terbanyak tahun 2018 - 2022**

```python
wind_direction_counts = data_meteorologi['arah terbanyak'].value_counts()

sns.set_palette("rainbow")

explode = [0.1 if i == wind_direction_counts.idxmax() else 0 for i in wind_direction_counts.index]

plt.figure(figsize=(8, 8))
plt.pie(wind_direction_counts, labels=wind_direction_counts.index, autopct='%1.1f%%', startangle=140, explode=explode, textprops={'fontsize': 11, 'color':'black' if wind_direction_counts.idxmax() else 'white'})
plt.title('\nFrekuensi Kemunculan Arah Angin Terbanyak 2018 - 2022\n')
plt.axis('equal')
plt.show()
```

**Menampilkan histogram untuk menampilkan distribusi curah hujan**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_meteorologi['curah hujan'], bins=20, kde=True, color='skyblue', edgecolor='black')

mean_rainfall = data_meteorologi['curah hujan'].mean()
plt.axvline(mean_rainfall, color='red', linestyle='dashed', linewidth=1, label=f'Rata-rata: {mean_rainfall:.2f} mm')

plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Frekuensi')
plt.title('\nHistogram Curah Hujan\n')
plt.legend()

plt.show()
```

## Timeseries Analysis

**Menganalisis Tren Temperatur Tahunan 2018 - 2022**

```python
plt.figure(figsize=(12, 6))
plt.plot(data_meteorologi['Tanggal'], data_meteorologi['Temperature max'], marker='o', color='red', label='Temperature Maksimum')
plt.plot(data_meteorologi['Tanggal'], data_meteorologi['Temperature rata2'], marker='o', color='orange', label='Temperature Rata-rata')
plt.plot(data_meteorologi['Tanggal'], data_meteorologi['Temperature min'], marker='o', color='green', label='Temperature Minimum')
plt.title('\nTren Temperatur Tahunan 2018 - 2022\n')
plt.xlabel('Tanggal')
plt.ylabel('Temperatur (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

tanggal_tinggi = data_meteorologi.loc[data_meteorologi['Temperature max'].idxmax()]['Tanggal']
suhu_tinggi = data_meteorologi['Temperature max'].max()

plt.annotate(f'     Maksimum: {suhu_tinggi} °C\n     {tanggal_tinggi.strftime("%Y-%m-%d")}',
             xy=(tanggal_tinggi, suhu_tinggi),
             xytext=(tanggal_tinggi + pd.DateOffset(days=5), suhu_tinggi - 2),
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=10,
             ha='left')

plt.show()
```

**Menganalisis Tren Temperatur Bulanan 2018 - 2022**

```python
data_meteorologi['Tanggal'] = pd.to_datetime(data_meteorologi['Tanggal'])

data_meteorologi['Bulan'] = data_meteorologi['Tanggal'].dt.month
data_meteorologi['Tahun'] = data_meteorologi['Tanggal'].dt.year

average_monthly_temperature = data_meteorologi.groupby(['Tahun', 'Bulan'])['Temperature rata2'].mean()

plt.figure(figsize=(12, 6))
average_monthly_temperature.plot(marker='o', color='black')

plt.title('\nTren Temperatur Bulanan (Rata-rata) 2018 - 2022\n')
plt.xlabel('Tanggal')
plt.ylabel('Temperatur Rata-rata (°C)')

ticks = range(0, len(average_monthly_temperature), 3)  # Label setiap tiga bulan
labels = [average_monthly_temperature.index[idx] for idx in ticks]  # Label yang sesuai dengan posisi xticks
plt.xticks(ticks=ticks, labels=labels, rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```
